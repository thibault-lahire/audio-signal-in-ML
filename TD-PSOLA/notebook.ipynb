```python
import os, sys, wave, struct

import numpy as np
import pyaudio
import pandas as pd
import matplotlib.pyplot as plt

from copy import deepcopy
from math import ceil

from scipy.io import wavfile
import mdct
from scipy.io.wavfile import write
```

```python
def load_music(file):
    return wave.open(file, 'rb')


def play_music(file, chunk = 1024):
    """
    Script from PyAudio doc
    """
    wf = load_music(file)
    p = pyaudio.PyAudio()
    stream = p.open(format=p.get_format_from_width(wf.getsampwidth()),
                    channels=wf.getnchannels(),
                    rate=wf.getframerate(),
                    output=True)
    data = wf.readframes(chunk)

    while data:
        stream.write(data)
        data = wf.readframes(chunk )

    stream.stop_stream()
    stream.close()
    p.terminate()
    
    
def plot_sound(data, times, name='default_name', save=False):
    plt.figure(figsize=(30, 4))
    plt.fill_between(times, data)
    plt.xlim(times[0], times[-1])
    plt.xlabel('time (s)')
    plt.ylabel('amplitude')
    if save:
        plt.savefig(name+'.png', dpi=100)
    plt.show()

    
def nextpow2(x):
    return ceil(np.log2(np.abs(x)))


def write_file(data, Fs, name):
    scaled = np.int16(data/np.max(np.abs(data)) * 32767)
    write(name+'.wav', Fs, scaled)
    return name+'.wav'

def listen_file(filename):
    current_path = os.getcwd()
    data_path = os.path.join(current_path, 'Sons')
    music = os.path.join(data_path, filename)
    play_music(music)
```

## 1 Extraction of the signal analysis marks

```python
def period(x, Fs, Pmin=1/300, Pmax=1/80, seuil=0.7):
    # [P, voiced] = period(x, Fs, Pmin, Pmax, seuil);
    # If voiced = 1, P is the period signal x expressed in number of samples
    # If voiced = 0, P is equal to 10ms.Fs
    
    x = x - np.mean(x)
    N = len(x)
    
    Nmin = np.ceil(Pmin*Fs).astype(int)
    Nmax = 1 + np.floor(Pmax*Fs).astype(int)
    Nmax = np.min([Nmax,N])
    
    Nfft = 2**int(nextpow2(2*N-1))
    X = np.fft.fft(x, n=Nfft)
    S = X*np.conj(X)/N
    r = np.real(np.fft.ifft(S))
    rmax = np.max(r[Nmin:Nmax])
    I = np.argmax(r[Nmin:Nmax])
    P = I + Nmin
    corr = (rmax/r[0])*(N/(N-P))
    voiced = corr > seuil
    
    if not(voiced):
        P = np.round(0.01*Fs)
        
    return P, voiced
```

```python
current_path = os.getcwd()
data_path = os.path.join(current_path, 'Sons')
filename = 'an_in_on.wav'
music = os.path.join(data_path, filename) 

wavefile = load_music(music)
print(wavefile.getparams())
play_music(music)
```

```python
Fs = int(wavefile.getframerate())
num_samples = int(wavefile.getnframes())
data = wavefile.readframes(num_samples)
data = struct.unpack('{n}h'.format(n=num_samples), data)
s = np.array(data)
timestep = 1/float(Fs)
times = np.arange(len(s))*timestep
plot_sound(s, times)
```

```python
def analysis_pitch_marks(s, Fs):
    A_1, A_2, A_3 = [], [], []
    A_1.append(int(0))
    A_2.append(False)
    A_3.append(int(Fs*10e-3))
    count = A_3[-1]
    while int(A_1[-1]+2.5*A_3[-1])<len(s):
        x = s[A_1[-1]:int(A_1[-1]+2.5*A_3[-1])]
        P_a_n, v_a_n = period(x, Fs)
        t_a_n = A_1[-1] + P_a_n
        A_1.append(int(t_a_n))
        A_2.append(v_a_n)
        A_3.append(int(P_a_n))
        count = A_3[-1] + int(P_a_n)
    A = np.vstack((A_1, A_2, A_3))
    return A
```

## 2 Synthesis and modification of the temporal and spectral scales

```python
A = analysis_pitch_marks(s, Fs)
B = np.zeros((2,A.shape[1]))
B[0] = A[0]
B[1] = np.arange(A.shape[1])
B = B.astype('int')
print(A.shape)
```

### 2.1 Signal synthesis

```python
def synthesis(s, Fs, A, B):
    y = np.zeros(int(B[0,-1] + A[2,int(B[1,-1])]))
    for k in range(1,len(B[0]-1)):
        x = s[(A[0,int(B[1,k])]-A[2,int(B[1,k])]):(A[0,int(B[1,k])]+A[2,int(B[1,k])])]
        w = np.hanning(len(x))
        x_w = x*w
        y[(int(B[0,k])-A[2,int(B[1,k])]):(int(B[0,k])+A[2,int(B[1,k])])] += x_w
    return y
```

```python
y = synthesis(s, Fs, A, B).astype('int')
write_file(y, Fs, 'Sons/synthesis')
listen_file('synthesis.wav')
```

### 2.2 Modification of the temporal scale

```python
def change_time_scale(alpha, A, Fs):
    B_1, B_2 = [], []
    B_1.append(int(1))
    B_2.append(int(1))
    while int(B_2[-1])<A.shape[1]-1:
        B_1.append(B_1[-1] + A[2, int(B_2[-1])])
        B_2.append(B_2[-1] + 1/alpha)
    B = np.vstack((B_1, B_2))
    B[1,-1] = A.shape[1]-1
    return B
```

```python
alpha = 3.8
B_time = change_time_scale(alpha, A, Fs)
y_time = synthesis(s, Fs, A, B_time)
write_file(y_time, Fs, 'Sons/time_change')
listen_file('time_change.wav')
```

### 2.3 Modification of the spectral scale

```python
def change_pitch_scale(beta, A, Fs):
    B_1, B_2 = [], []
    B_1.append(int(1))
    B_2.append(int(1))
    while int(B_2[-1])<A.shape[1]-1:
        if A[1,int(B_2[-1])]:
            scale = 1/beta
        else:
            scale = 1
        B_1.append(B_1[-1] + scale*A[2, int(B_2[-1])])
        B_2.append(B_2[-1] + scale)
    B = np.vstack((B_1, B_2))
    B[1,-1] = A.shape[1]-1
    return B
```

```python
beta = 3.8
B_pitch = change_pitch_scale(beta, A, Fs)
y_pitch = synthesis(s, Fs, A, B_pitch)
write_file(y_pitch, Fs, 'Sons/change_pitch')
listen_file('change_pitch.wav')
```

### 2.4 Joint modification of the temporal and spectral scales

```python
def change_both_scales(alpha, beta, A, Fs):
    B_1, B_2 = [], []
    B_1.append(int(1))
    B_2.append(int(1))
    while int(B_2[-1])<A.shape[1]-1:
        if A[1,int(B_2[-1])]:
            scale = 1/beta
            B_1.append(B_1[-1] + scale*A[2, int(B_2[-1])])
            B_2.append(B_2[-1] + scale/alpha)
        else:
            B_1.append(B_1[-1] + A[2, int(B_2[-1])])
            B_2.append(B_2[-1] + 1/alpha)
    B = np.vstack((B_1, B_2))
    B[1,-1] = A.shape[1]-1
    return B
```

```python
alpha = 3.8
beta = 3.8
B_both = change_both_scales(alpha, beta, A, Fs)
y_both = synthesis(s, Fs, A, B_both)
write_file(y_both, Fs, 'Sons/both')
listen_file('both.wav')
```
